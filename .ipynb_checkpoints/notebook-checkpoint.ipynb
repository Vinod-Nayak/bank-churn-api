# Data Preparation and Data Cleaning

1. Standardize column names

```python
df.columns = df.columns.str.lower().str.strip()
print("Column names standardized:")
print(df.columns.tolist())
```

2. Handling Missing Values:

```python
df.isnull().sum()
```

3. Encode categorical values:

```python
# Remove identifier column if present
if 'customer_id' in df.columns:
    df.drop(columns=['customer_id'], inplace=True)
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df = pd.get_dummies(df, columns=['country'], drop_first=True)
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip()
print("Column names standardized:")
print(df.columns.tolist())
```

4. Ensure Boolean dummies are numeric

```python
for col in ['country_germany', 'country_spain']:
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)

print(df.dtypes)
```

```python
print("\nChurn value counts:")
print(df['churn'].value_counts())
```

```python
df.to_csv("data/cleaned_bank_churn.csv", index=False)
print("Cleaned dataset saved to data/cleaned_bank_churn.csv")
```

# Exploratory Data Analysis (EDA)

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("data/cleaned_bank_churn.csv")

# Target variable distribution
sns.countplot(x='churn', data=df)
plt.title('Churn Distribution (0 = Stayed, 1 = Left)')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()

print("Churn ratio:")
print(df['churn'].value_counts(normalize=True))

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()

# Churn by Gender
sns.barplot(x='gender', y='churn', data=df, estimator='mean')
plt.title('Churn Rate by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Churn Rate')
plt.show()

# Churn by Country
country_cols = ['country_germany', 'country_spain']
for col in country_cols:
    churn_rate = df.groupby(col)['churn'].mean()
    print(f"\nChurn Rate by {col}:")
    print(churn_rate)
```

Numeric Feature Relationships

```python
# Age vs Churn
sns.boxplot(x='churn', y='age', data=df)
plt.title('Age Distribution by Churn Status')
plt.show()

# Balance vs Churn
sns.boxplot(x='churn', y='balance', data=df)
plt.title('Balance Distribution by Churn Status')
plt.show()

# Credit Score vs Churn
sns.boxplot(x='churn', y='credit_score', data=df)
plt.title('Credit Score Distribution by Churn Status')
plt.show()

# Active Member vs Churn
sns.barplot(x='active_member', y='churn', data=df, estimator='mean')
plt.title('Churn Rate by Active Membership')
plt.xlabel('Active Member (0 = No, 1 = Yes)')
plt.ylabel('Average Churn Rate')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Split into features & target
X = df.drop('churn', axis=1)
y = df['churn']

# Scale numeric data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train model for feature importance
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Feature importance
importances = rf.feature_importances_
features = X.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)


print(importance_df)

# Plot top 10 important features
plt.figure(figsize=(10, 5))
sns.barplot(x='Importance', y='Feature', data=importance_df.head(10), palette='viridis')
plt.title('Top 10 Feature Importances (Random Forest)')
plt.show()
```

```python
# ==========================================
#  MODEL SELECTION AND PARAMETER TUNING
# ==========================================
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("data/cleaned_bank_churn.csv")

# Split features and target
X = df.drop('churn', axis=1)
y = df['churn']

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# ------------------------------------------
# 1 Logistic Regression (Baseline)
# ------------------------------------------
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print(" Logistic Regression Results:")
print(classification_report(y_test, y_pred_lr))

# ------------------------------------------
# 2 Random Forest Classifier
# ------------------------------------------
rf = RandomForestClassifier(random_state=42)
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5],
}
rf_grid = GridSearchCV(rf, rf_params, cv=3, scoring='f1', n_jobs=-1, verbose=1)
rf_grid.fit(X_train, y_train)

best_rf = rf_grid.best_estimator_
y_pred_rf = best_rf.predict(X_test)

print(" Random Forest Results:")
print("Best Parameters:", rf_grid.best_params_)
print(classification_report(y_test, y_pred_rf))

# ------------------------------------------
# 3 Gradient Boosting / XGBoost
# ------------------------------------------
xgb_clf = xgb.XGBClassifier(eval_metric='logloss', random_state=42)
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.8, 1.0]
}

xgb_grid = GridSearchCV(xgb_clf, xgb_params, cv=3, scoring='f1', n_jobs=-1, verbose=1)
xgb_grid.fit(X_train, y_train)

best_xgb = xgb_grid.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)

print(" XGBoost Results:")
print("Best Parameters:", xgb_grid.best_params_)
print(classification_report(y_test, y_pred_xgb))

# ------------------------------------------
# 4 Compare Models
# ------------------------------------------
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_xgb)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_lr),
        f1_score(y_test, y_pred_rf),
        f1_score(y_test, y_pred_xgb)
    ],
    'ROC-AUC': [
        roc_auc_score(y_test, log_reg.predict_proba(X_test)[:,1]),
        roc_auc_score(y_test, best_rf.predict_proba(X_test)[:,1]),
        roc_auc_score(y_test, best_xgb.predict_proba(X_test)[:,1])
    ]
})

print("\n Model Comparison:")
print(results.sort_values(by='F1 Score', ascending=False))

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='F1 Score', y='Model', data=results.sort_values(by='F1 Score', ascending=False), palette='mako')
plt.title('Model Performance Comparison')
plt.show()
```
